```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_conversion_by_segment(df, seg_col, target='y_num',
                               title=None, figsize=(6,4)):

    rates = (
        df
        .groupby(seg_col)[target]
        .mean()
        .reset_index()
        .sort_values(target, ascending=False)
    )

    plt.figure(figsize=figsize)
    sns.barplot(data=rates, x=seg_col, y=target, palette = "Oranges",edgecolor="black")
    plt.xticks(rotation=45, ha='right')
    plt.ylim(0, rates[target].max()*1.1)
    plt.ylabel("Subscription Rate")
    plt.xlabel(seg_col.replace('_',' ').title())
    if title:
        plt.title(title)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# 1) Compute Pearson’s r for numeric features
numeric_cols = ['age','balance','day','campaign','pdays','previous','duration']
pearsons = (
    df[numeric_cols + ['y_num']]
      .corr()['y_num']
      .drop('y_num')
      .abs()
      .sort_values(ascending=False)
)

# 2) Compute Cramér’s V for categorical features
def cramers_v(x, y):
    conf = pd.crosstab(x, y)
    chi2, _, _, _ = chi2_contingency(conf)
    n = conf.values.sum()
    r, k = conf.shape
    return np.sqrt(chi2 / (n * (min(r, k) - 1)))

categorical_cols = ['job','marital','education','default',
                    'housing','loan','contact','month','poutcome']
cat_corr = pd.Series({
    col: cramers_v(df[col], df['y_num'])
    for col in categorical_cols
}).sort_values(ascending=False)

# 3) Take top‑3 of each
top_num = pearsons.head(3)
top_cat = cat_corr.head(3)

# 4) Plot side by side
fig, axes = plt.subplots(1, 2, figsize=(10,4))

sns.barplot(
    x=top_num.values, 
    y=top_num.index, 
    ax=axes[0], 
    palette="Oranges_r"
)
axes[0].set_title("Top‑3 Numeric → Subscription")
axes[0].set_xlabel("Correlation for Numeric")

sns.barplot(
    x=top_cat.values, 
    y=top_cat.index, 
    ax=axes[1], 
    palette="Oranges_r"
)
axes[1].set_title("Top‑3 Categorical → Subscription")
axes[1].set_xlabel("Correlation for Categories")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# 1) Define a function to compute Cramér’s V between two categorical variables
def cramers_v(x, y):
    """Cramér’s V statistic for categorical–categorical association."""
    # build contingency table
    confusion = pd.crosstab(x, y)
    chi2, p, dof, expected = chi2_contingency(confusion)
    n = confusion.sum().sum()
    phi2 = chi2 / n
    r, k = confusion.shape
    # bias correction
    phi2corr = max(0, phi2 - ((k - 1)*(r - 1)) / (n - 1))
    rcorr = r - ((r - 1)**2)/(n - 1)
    kcorr = k - ((k - 1)**2)/(n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

# 2) List your categorical features (excluding the target)
cat_cols = [
    'job', 'marital', 'education', 'default',
    'housing', 'loan', 'contact', 'month', 'poutcome'
]

# 3) Build an empty DataFrame to hold the pairwise Cramér’s V values
cramer_mat = pd.DataFrame(index=cat_cols, columns=cat_cols, dtype=float)

# 4) Fill it in
for i in cat_cols:
    for j in cat_cols:
        cramer_mat.loc[i, j] = cramers_v(df[i], df[j])

# 5) Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    cramer_mat.astype(float),
    annot=True,
    fmt=".2f",
    cmap="Oranges",
    cbar_kws={"label": "Cramér’s V"}
)
plt.title("Cramér’s V Heatmap (Categorical ↔ Categorical)")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make sure we’re in “inline” mode if you’re in a notebook:
# %matplotlib inline

# 1) pick one source of truth
df_plot = df_clean.copy()  

numeric_cols = ['age','balance','day','campaign','pdays','previous','duration']

# 2) Melt  
melted = df_plot[numeric_cols + ['y_num']].melt(
    id_vars='y_num',
    var_name='feature',
    value_name='value'
)

# 3) Plot
plt.figure(figsize=(10, 6))
ax = sns.boxplot(
    data=melted,
    y='feature',
    x='value',
    hue='y_num',
    palette={0:'lightgray', 1:'steelblue'},
    fliersize=2,
    whis=1.5
)

# use symlog so that 0’s show up, with a linear region [-1,1]
ax.set_xscale('symlog', linthresh=1)
ax.set_xlabel("Value (symlog scale)")
ax.set_ylabel("")
ax.legend(title="Subscribed", loc='lower right')
ax.set_title("Feature Distributions by Subscription (1.5×IQR)")

# 4) Annotate outlier counts
for i, feat in enumerate(numeric_cols):
    vals = df_plot[feat].dropna().astype(float)
    Q1, Q3 = np.percentile(vals, [25, 75])
    iqr = Q3 - Q1
    lower, upper = Q1 - 1.5*iqr, Q3 + 1.5*iqr
    n_out = ((vals < lower) | (vals > upper)).sum()
    # text just outside the positive whisker
    xpos = upper if upper>0 else vals.max()
    ax.text(
        x=xpos * 1.1,
        y=i,
        s=f"{n_out} outliers",
        va='center',
        ha='left',
        color='darkred',
        fontsize=9
    )

plt.tight_layout()
plt.show()
```
